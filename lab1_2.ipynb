# 4. Информация о корреляции признаков

```python
data1.corr()
```

На основе корреляционной матрицы можно сделать следующие выводы:
1. Целевой признак наиболее сильно коррелирует с Флавоноидами (-0.8), с Фенолами (-0.7), разбавленностью вина (-0.7), с коэффициентом оттенка (-0.6), с сортом вина (-0.6)
2. Целевой признак от части коррелирует со щелочностью осадка (0.5)
3. Целевой признак слабо коррелирует с коэффициентом яблочной кислоты (0.4), коэффициентом осадка (0.04), проантоцианидами (-0.4)
4. Флаваноиды и коэффициент фенолов сильно коррелируют между собой (0.8)

По умолчанию при построении матрицы используется коэффициент корреляции Пирсона, построим корреляционные матрицы на основе коэффициентов Кендалла и Спирмена.

```python
data1.corr(method='pearson')
```

```python
data1.corr(method='kendall')
```

```python
data1.corr(method='spearman')
```

Для анализа большого колчества признаков удобнее использовать цветовой анализ коррелиционной карты. Будем использовать "тепловую карту".

```python
sns.heatmap(data1.corr())
```

```python
# добавим вывод значений в ячейках
sns.heatmap(data1.corr(), annot=True, fmt='.3f')
```

```python
# изменение цветовой гаммы
sns.heatmap(data1.corr(), cmap='YlGnBu', annot=True, fmt='.3f')
```

```python
# уберем числовые значения
sns.heatmap(data1.corr(), cmap='YlGnBu', fmt='.3f')
```

```python
# треугольный вариант матрицы
mask = np.zeros_like(data1.corr(), dtype=np.bool)
mask[np.tril_indices_from(mask)] = True
sns.heatmap(data1.corr(), mask=mask, fmt='.3f')
```

```python
fig, ax = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15,5))
sns.heatmap(data1.corr(method='pearson'), ax=ax[0], fmt='.2f')
sns.heatmap(data1.corr(method='kendall'), ax=ax[1], fmt='.2f')
sns.heatmap(data1.corr(method='spearman'), ax=ax[2], fmt='.2f')
fig.suptitle('Корреляционные матрицы, построенные различными методами')
ax[0].title.set_text('Pearson')
ax[1].title.set_text('Kendall')
ax[2].title.set_text('Spearman')
```
